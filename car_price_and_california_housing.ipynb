# **Linear Regression**

```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
from sklearn import datasets
dir(datasets)
```

**fetch_california_housing**

```python
# Data Preparation

data = fetch_california_housing()
data
data.keys()
# print(data.DESCR)
df = pd.DataFrame(data.data,columns = data.feature_names)
df
data.target_names[0]
df[data.target_names[0]] = data.target
df
```

**Model Training**

3 steps to train
1) divide the data into input and output
2)  split data into training and testing

```python
# 3 steps to train
# 1) divide the data into input and output

X = df.drop("MedHouseVal",axis = 1)
y = df['MedHouseVal']
```

```python
# ML rules --> 100% --> 70%(training) , 30%(testing)
# split data into training and testing
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3,random_state=42)
Xtrain
```

```python
# model traing
model = LinearRegression()
model.fit(Xtrain,ytrain)
```

****

```python
print("Intercept : ",model.intercept_)
print("Slope : ",model.coef_)
```

**Model Evaluation**

```python
ypred = model.predict(Xtest)
ypred
```

```python
print("Metrics : ")
r2 = r2_score(ytest,ypred)
print("R2 : ",r2)
N = df.shape[0]
P = df.shape[1]
adj_r2 = 1 - (((1 - r2**2)*(N-1))/N-P-1)
print("Adj r2 : ",adj_r2)
print("MSE : ",mean_squared_error(ytest,ypred))
print("MAE : ",mean_absolute_error(ytest,ypred))
```

```python
print("Training eff : ",model.score(Xtrain,ytrain))
print("Training eff : ",model.score(Xtest,ytest))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Xtrain = scaler.fit_transform(Xtrain)
Xtest = scaler.transform(Xtest) # to avoid data leakage we should use just transform -- when we use fit the model will learn -- hence to avoid the model to see the test we wont use fit
```

# **car_price_prediction**

```python
df = pd.read_csv("/content/car_price_prediction.csv")
df.head()
```

```python
df.info()
```

**cleaning**

```python
df.isnull().sum()
```

```python
df = df.drop(["ID","Doors"],axis = 1)
```

```python
df['Levy'].unique()
```

```python
# clean Levy
df['Levy'] = df['Levy'].replace("-",np.nan)
df['Levy'] = df['Levy'].astype(float)
avg = round(df['Levy'].mean())
df['Levy'] = df['Levy'].fillna(avg)
df['Levy'] = df['Levy'].astype(int)
```

```python
df.info()
```

```python
df['Manufacturer'].unique()
mode = df['Manufacturer'].mode()[0]
df['Manufacturer'] =df['Manufacturer'].str.replace('სხვა',mode)

df['Manufacturer'].unique()
```

```python
df['Engine volume'].unique()
df['Engine volume'] = df['Engine volume'].str.replace(' Turbo','')
df['Engine volume'] = df['Engine volume'].astype(float)
df['Engine volume'].unique()
```

```python
df['Mileage'].unique()
df['Mileage'] = df['Mileage'].str.replace(' km','')
df['Mileage'] = df['Mileage'].astype(float)
df['Mileage'].unique()
```

```python
df['Airbags'].unique()
```

```python
df.drop_duplicates()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for col in df.columns:
  if df[col].dtype == object:
    df[col] = encoder.fit_transform(df[col])

df.head()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot = True)
```

```python
X = df.drop("Price",axis = 1)
y = df["Price"]
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size = 0.3,random_state = 42)
```

```python
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# Xtrain = scaler.fit_transform(Xtrain)
# Xtest = scaler.transform(Xtest)
```

```python
# model traing
model = LinearRegression()
model.fit(Xtrain,ytrain)
```

```python
ypred = model.predict(Xtest)
ypred
```

```python
print("Training eff : ",model.score(Xtrain,ytrain))
print("Testing eff : ",model.score(Xtest,ytest))
```

```python
print("Metrics : ")
r2 = r2_score(ytest,ypred)
print("R2 : ",r2)
N = df.shape[0]
P = df.shape[1]
adj_r2 = 1 - (((1 - r2**2)*(N-1))/N-P-1)
print("Adj r2 : ",adj_r2)
print("MSE : ",mean_squared_error(ytest,ypred))
print("MAE : ",mean_absolute_error(ytest,ypred))
```
